```python
pares = []
impares = []
```

```python
N = int(input())
```

```python
for i in range(N):
    num = int(input())
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)
```

```python
pares.sort()
impares.sort(reverse=True)
```

```python
for numero in pares:
    print(numero)
for numero in impares:
    print(numero)
```
